## Zadanie 3

W tym zadaniu zdefiniowano funkcję $f(x)=x^2+5$ oraz sporządzono jej wykres na 3 zadanych przedziałach:


1. $x>-1\;oraz\;x<1$
2. $x>-6\;oraz\;x<6$
3. $x>0\;\;\;\;oraz\;x<5$

```python
import matplotlib.pyplot as plt
import numpy as np

# definicja funkcji f
f = lambda x : (x**2) + 5

# wektory odpowiednich przedziałów
x1 = np.linspace(-1,1)
x2 = np.linspace(-6,6)
x3 = np.linspace(-0,5)
```

**Przedział 1**

```python
plt.plot(x1, f(x1), label="$f (x)=x^2+5$")
plt.title("Wykres funkcji $f (x)=x^2+5$ na przedziale $x∈(-1,1)$")
plt.legend()
plt.xlabel('Oś x')
plt.ylabel('Oś y')
plt.show()
```

**Przedział 2**

```python
plt.plot(x2, f(x2), label="$f (x)=x^2+5$")
plt.title("Wykres funkcji $f (x)=x^2+5$ na przedziale $x∈(-6,6)$")
plt.legend()
plt.xlabel('Oś x')
plt.ylabel('Oś y')
plt.show()
```

**Przedział 3**

```python
plt.plot(x3, f(x3), label="$f (x)=x^2+5$")
plt.title("Wykres funkcji $f (x)=x^2+5$ na przedziale $x∈(0,5)$")
plt.legend()
plt.xlabel('Oś x')
plt.ylabel('Oś y')
plt.show()
```
