```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../Original Data/tb.csv')
df
```

```python
# wywalenie wierszy wypeÅ‚nionych jedynie wartoÅ›ciami NaN oraz zamiana NaN na 0
df = df.dropna(thresh=3)
df = df.fillna(0)
df
```

```python
# utworzenie zestawu danych gdzie obserwacje = lata

df1 = df.drop(columns=['iso2'])
df1 = df1.groupby(['year'], as_index=False).sum()
df1.to_csv('../Analysis Data/tb_in_years.csv')
df1
```

```python
# utworzenie zestawu z pominiÄ™ciem przedziaÅ‚Ã³w wiekowych
df1['new_sp_male'] = df1[list(df1.columns)[2:12]].sum(axis=1)
df1['new_sp_female'] = df1[list(df1.columns)[12:22]].sum(axis=1)
df1 = df1.drop(columns=list(df1.columns)[2:22])
df1.to_csv('../Analysis Data/tb_in_years_without_age.csv')
df1
```

```python
# wizualizacja nowych przypadkÃ³w u pÅ‚ci mÄ™skiej

plt.bar(df1['year'], df1['new_sp_male'])
plt.show()
```

```python
# utowrzenie zestawu z pominiÄ™ciem pÅ‚ci
df1['all_new_sp'] = df1[list(df1.columns)].sum(axis=1)
df1 = df1.drop(columns=list(df1.columns)[1:4])
df1.to_csv('../Analysis Data/tb_in_years_without_age_and_sex.csv')
df1
```

```python
# wizualizacja nowych przypadkÃ³w w poszczegÃ³lnych latach

plt.bar(df1['year'], df1['all_new_sp'])
plt.show()
```
